```python
df_work = df[["Datetime", "Total_Occupancy"]].copy()
df_work.head()
```

```python
df_work["Datetime"] = pd.to_datetime(df_work["Datetime"])
df_work.dtypes
```

```python
df_work = df_work.sort_values("Datetime")
df_work.head()
```

```python
df_work["Total_Occupancy"].describe()
```

```python
high_occ_threshold = df_work["Total_Occupancy"].quantile(0.90)
high_occ_threshold
```

```python
df_work["High_Occupancy"] = df_work["Total_Occupancy"] >= high_occ_threshold
df_work.head()
```

```python
high_occ_rate = df_work["High_Occupancy"].mean()
high_occ_rate
```

```python
df_work["High_Run_ID"] = (
    df_work["High_Occupancy"]
    .ne(df_work["High_Occupancy"].shift())
    .cumsum()
)

df_work.head()
```

```python
high_runs = (
    df_work[df_work["High_Occupancy"]]
    .groupby("High_Run_ID")
    .size()
)

high_runs.describe()
```

```python
longest_run = high_runs.max() if not high_runs.empty else 0
longest_run
```

```python
if high_occ_rate >= 0.20 and longest_run >= 3:
    decision = "Increase seating"
elif longest_run >= 3:
    decision = "Extend operating hours"
elif high_occ_rate > 0:
    decision = "Monitor only"
else:
    decision = "No action needed"

decision
```

```python
df_work["Hour"] = df_work["Datetime"].dt.hour

def time_bucket(hour):
    if 8 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    else:
        return "Evening"

df_work["Time_Bucket"] = df_work["Hour"].apply(time_bucket)
df_work[["Datetime", "Hour", "Time_Bucket"]].head()
```

```python
bucket_pressure = (
    df_work.groupby("Time_Bucket")["High_Occupancy"]
    .mean()
)

bucket_pressure
```

```python
most_stressed_bucket = bucket_pressure.idxmax()
most_stressed_bucket
```
